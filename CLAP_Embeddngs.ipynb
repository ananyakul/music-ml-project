```python
!pip install transformers[torch] torchaudio --upgrade
from transformers import AutoProcessor, ClapModel
from torch.nn.functional import cosine_similarity
from google.colab import drive
import torch
import torchaudio
```

```python
model = ClapModel.from_pretrained("laion/clap-htsat-unfused")
processor = AutoProcessor.from_pretrained("laion/clap-htsat-unfused")


drive.mount('/content/drive')
audio_path = "/content/drive/My Drive/MIT-Education/Junior Year/Music ML/Poolside-MusicLM.wav"
audio_sample, sample_rate = torchaudio.load(audio_path)
input_text = ["Sound of a dog", "Sound of vacuum cleaner"]
inputs = processor(text=input_text, audios=audio_sample, sampling_rate=48000, return_tensors="pt", padding=True)
```

```python
outputs = model(**inputs)
logits_per_audio = outputs.logits_per_audio  # This is the audio-text similarity score
probs = logits_per_audio.softmax(dim=-1)
print("Similarity Scores:", logits_per_audio)
print("Probabilities:", probs)
```
